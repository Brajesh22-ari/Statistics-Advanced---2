```python
1.: What is hypothesis testing in statistics?
ANS.Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data.
Key Steps in Hypothesis Testing

Formulate Hypotheses

Null Hypothesis (H₀): Assumes no effect, no difference, or status quo.
Example: "The new drug has no effect compared to the old drug."

Alternative Hypothesis (H₁ or Ha): States there is an effect or difference.
Example: "The new drug is more effective than the old drug."

Choose a Significance Level (α)

Commonly 0.05 (5%).

It represents the probability of rejecting H₀ when it is actually true (Type I error).

Select the Test & Collect Data

Depending on the problem: z-test, t-test, chi-square test, ANOVA, etc.

Compute Test Statistic & P-value
Test statistic: Measures how far the sample result is from what is expected under H₀.

P-value: Probability of getting results at least as extreme as the sample result, assuming H₀ is true.

Decision Rule

If p-value ≤ α, reject H₀ (evidence supports H₁).

If p-value > α, fail to reject H₀ (not enough evidence against H₀).
Example

Suppose a company claims the average lifetime of its batteries is 100 hours.

H₀: μ = 100

H₁: μ ≠ 100

You test 30 batteries, compute the sample mean, and run a t-test.

If p-value = 0.02 and α = 0.05 → Reject H₀ (evidence that mean ≠ 100).
```

```python
2.What is the null hypothesis, and how does it differ from the alternative
hypothesis?
ANS.1. Null Hypothesis (H₀)

It is the default assumption or the "no effect / no difference" statement.

It represents the idea that nothing new or unusual is happening.

The goal of the test is to see if we have enough evidence to reject H₀.

 Examples:

A coin is fair → H₀: p = 0.5

Average height of students = 170 cm → H₀: μ = 170

A new drug is no better than the old one → H₀: μ₁ = μ₂
2. Alternative Hypothesis (H₁ or Ha)

It is the statement that contradicts H₀.

It represents the claim we want to test or prove (effect, difference, or change exists).

If evidence strongly contradicts H₀, we accept H₁.

 Examples:

The coin is biased → H₁: p ≠ 0.5

Average height ≠ 170 cm → H₁: μ ≠ 170

The new drug is better → H₁: μ₁ > μ₂
| Aspect     | Null Hypothesis (H₀)                        | Alternative Hypothesis (H₁)                 |
| ---------- | ------------------------------------------- | ------------------------------------------- |
| Meaning    | No effect, no difference, status quo        | Effect, difference, or change               |
| Assumption | Accepted as true until proven otherwise     | Considered if data provides enough evidence |
| Symbol     | “=” sign is always included (e.g., μ = 170) | Uses ≠, >, or < (e.g., μ ≠ 170, μ > 170)    |
| Goal       | Try to reject it with evidence              | Try to support it if H₀ is rejected         |
```

```python
3.: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
ANS..Significance Level (α)

The significance level (α) is a threshold we set before conducting a hypothesis test.
It represents the probability of making a Type I error → rejecting the null hypothesis (H₀) when it is actually true.
Common choices:

α = 0.05 (5%) → most widely used

α = 0.01 (1%) → stricter, used in medical/critical research

α = 0.10 (10%) → less strict, used in exploratory studies
Role in Hypothesis Testing

Compare p-value to α

If p ≤ α → Reject H₀ (enough evidence for H₁).

If p > α → Fail to reject H₀ (not enough evidence against H₀).

Sets the “risk tolerance”

Smaller α → stricter test, reduces false positives but increases chance of false negatives.

Larger α → more lenient, increases chance of detecting an effect but also increases false positives.
Example

A company claims the average weight of chocolate bars = 50g.

H₀: μ = 50

H₁: μ ≠ 50

Significance level: α = 0.05

After testing, you calculate p-value = 0.03.

Since 0.03 < 0.05 → Reject H₀.

Conclusion: Evidence suggests the average weight is not 50g.

 In short:
 The significance level (α) is the cutoff point that decides whether the evidence from the sample is strong enough to reject the null hypothesis. It controls the risk of making a wrong decision (Type I error).
```

```python
4.: What are Type I and Type II errors? Give examples of each?
ANS. Type I Error (False Positive)

Occurs when we reject the null hypothesis (H₀) even though it is true.

In other words → we think something is happening when actually nothing is happening.

Probability of this error = α (significance level).

 Examples

A medical test says a healthy person has a disease.

A court convicts an innocent person.

A company concludes their new drug is effective, but in reality, it isn’t.
Type II Error (False Negative)

Occurs when we fail to reject the null hypothesis (H₀) even though it is false.

In other words → we miss a real effect or difference.

Probability of this error = β.

The ability to avoid Type II error is called power of the test = (1 - β).

 Examples

A medical test fails to detect a disease in a sick person.

A court lets a guilty person go free.

A company concludes their new drug has no effect, when in reality it does.
| Decision ↓ / Reality → | H₀ is True                          | H₀ is False                          |
| ---------------------- | ----------------------------------- | ------------------------------------ |
| Reject H₀          | Type I Error (False Positive) |  Correct (detects real effect)      |
| Fail to Reject H₀  |  Correct (no effect)               |Type II Error (False Negative) |
Type I error = false alarm (detecting something that isn’t there).

Type II error = missed detection (failing to see something that is there).
```

```python
5.: What is the difference between a Z-test and a T-test? Explain when to use
each.
ANS..Z-Test

Used when:

Population standard deviation (σ) is known.

Sample size is large (n ≥ 30).

Assumes data is normally distributed.

Based on the standard normal distribution (Z-distribution).

 Example:
A manufacturer claims that the average weight of sugar packets is 1 kg with σ = 0.05 kg known. You take a sample of 50 packets and want to test if the mean = 1 kg.
 Use Z-test (since σ is known and n ≥ 30).
 T-Test

Used when:

Population standard deviation (σ) is unknown.

Sample size is small (n < 30).

Based on the Student’s t-distribution, which has fatter tails (to handle more variability in small samples).

As sample size grows, the t-distribution approaches the Z-distribution.

 Example:
You want to test if the average marks of a class of 20 students = 70, but the population σ is unknown.
 Use T-test.
 | Feature             | Z-Test          | **T-Test**                  |
| ------------------- | -------------------- | --------------------------- |
| Population σ known? | Yes                  | No (estimated using sample) |
| Sample size (n)     | Large (≥ 30)         | Small (< 30)                |
| Distribution used   | Standard normal (Z)  | Student’s t                 |
| Variability         | Less (since σ known) | More (since σ estimated)    |
In short:

Use Z-test when σ is known and n is large.

Use T-test when σ is unknown and/or n is small.
```

```python
7.: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)

ANS.Let’s implement hypothesis testing using Z-statistics for the given sample dataset.

We’ll test the claim:

Null Hypothesis (H₀): μ = 50 (population mean = 50)

Alternative Hypothesis (H₁): μ ≠ 50 (population mean is not equal to 50)

We’ll use α = 0.05 (5% significance level).
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print(f"Sample Mean = {sample_mean:.3f}")
print(f"Sample Std Dev = {sample_std:.3f}")
print(f"Sample Size = {n}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Evidence suggests the population mean is different from 50.")
else:
    print("Fail to Reject H0: Not enough evidence to say the mean is different from 50.")
How it works

Compute sample mean & std.

Calculate Z-statistic:

𝑍
=
𝑥
ˉ
−
𝜇
0
𝑠
/
𝑛
Z=
s/
n
	​

x
ˉ
−μ
0
	​

	​


Compute p-value from standard normal distribution.

Compare p-value with α = 0.05.
```

```python
8.Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)
ANS... Python script that simulates data from a normal distribution, calculates the 95% confidence interval (CI) for its mean, and plots the data distribution using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # population mean and std dev
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Step 3: Calculate 95% confidence interval
alpha = 0.05
df = n - 1  # degrees of freedom
t_critical = stats.t.ppf(1 - alpha/2, df)  # t critical value
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean = {sample_mean:.3f}")
print(f"95% Confidence Interval = ({ci_lower:.3f}, {ci_upper:.3f})")

# Step 5: Plot histogram of data
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"Upper 95% CI = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
Generates 100 random samples from a normal distribution with μ = 50, σ = 5.

Computes sample mean and 95% confidence interval using the t-distribution (since σ is estimated from sample).

Plots histogram of the data and shows vertical lines for mean & CI
```

```python
9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

ANS.. Python implementation that calculates Z-scores for a dataset and then visualizes them using a histogram.
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores and plot histogram
def zscore_standardization(data):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std

    # Step 2: Compute Z-scores
    z_scores = (data - mean) / std

    # Step 3: Print first few Z-scores
    print("First 10 Z-scores:", z_scores[:10])

    # Step 4: Plot histogram of standardized data
    plt.hist(z_scores, bins=15, color='lightblue', edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (0)")
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    return z_scores

# Example dataset
data = np.random.normal(100, 15, 200)  # mean=100, std=15, size=200

# Call the function
z_scores = zscore_standardization(data)
Explanation

Z-score formula:

𝑍
=
𝑋
−
𝜇
𝜎
Z=
σ
X−μ
	​


Where:

𝑋
X = data value

𝜇
μ = mean of dataset

𝜎
σ = standard deviation

Interpretation:

Z-score tells how many standard deviations a data point is away from the mean.

Example:

Z = 0 → value is exactly at the mean.

Z = +1 → value is 1 standard deviation above mean.

Z = -2 → value is 2 standard deviations below mean.

After standardization:

Data has mean = 0 and std = 1.

Useful for comparing values across different scales.
```
